```python
import pandas as pd

train_data = pd.read_csv('Training-set.csv')

train_data.tail(4)
```

```python
# 筛选出t = 0的数据
df = pd.DataFrame(train_data)
df_0 = df[[each == 0 for each in df['t']]]
df_0
```

```python
for col in df_0.columns[1:]:
    print("该列数据的均值%.6f" %df_0[col].mean()) # 计算t == 0列的均值
```

```python
# 计算t == 1的数据
df = pd.DataFrame(train_data)
df_1 = df[[each == 1 for each in df['t']]]
df_1
```

```python
for col in df_1.columns[1:]:
    print("该列平均值%.6f" %df_1[col].mean()) # 计算t == 1列的平均值
```
